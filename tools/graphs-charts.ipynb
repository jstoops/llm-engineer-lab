Useful Graphs and Charts
========================

# Dependencies

```python
# imports

import matplotlib.pyplot as plt
```

# Setup

```python
%matplotlib inline
```

```python
ml_models = ["Constant", "Feature + LR", "Random Forest", "Human", "GPT-4o-mini", "Fine-Tuned Llama 3.1 Base 8B 4-bit"]
price_error = [145, 139, 96, 127, 79, 48]

# Bar chart by ML Model
plt.figure(figsize=(15, 6))
plt.bar(ml_models, price_error, color="skyblue")
plt.title('Average absolute prediction error from our models')
plt.xlabel('Models')
plt.ylabel('Price $')

plt.xticks(rotation=30, ha='right')

# Add value labels on top of each bar
for i, v in enumerate(price_error):
    plt.text(i, v, f"${v:,}", ha='center', va='bottom')

# Display the chart
plt.show()
```

# Bar Chart

```python
title = 'Title of bar chart'
bar_color = 'skyblue'
bars = ["Bar 1", "Bar 2", "Bar 3"]
bars_label = 'Bar labels'
data = [123, 321, 42]
data_label = 'Units label'

# Bar chart
plt.figure(figsize=(15, 6))
plt.bar(bars, data, color=bar_color)
plt.title(title)
plt.xlabel(bars_label)
plt.ylabel(data_label)

plt.xticks(rotation=30, ha='right')

# Add value labels on top of each bar
for i, v in enumerate(data):
    plt.text(i, v, f"{v:,}", ha='center', va='bottom')

# Display the chart
plt.show()
```

# Scatter Graph

```python
title = 'Title of scatter graph'
correlation_bar_color = 'deepskyblue'
dataset_1 = [123, 321, 42]
dataset_1_label = 'Dataset 1 Label'
dataset_2 = [324, 12, 54]
dataset_2_label = 'Dataset 2 Label'
datapoint_colors = 'red'
max_val = max(max(dataset_1), max(dataset_2))

# Scatter graph
plt.figure(figsize=(12, 8))
plt.plot([0, max_val], [0, max_val], color=correlation_bar_color, lw=2, alpha=0.6)
plt.scatter(dataset_1, dataset_2, s=3, c=datapoint_colors)
plt.xlabel(dataset_1_label)
plt.ylabel(dataset_2_label)
plt.xlim(0, max_val)
plt.ylim(0, max_val)
plt.title(title)

# Display the chart
plt.show()
```
